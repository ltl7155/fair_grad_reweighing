# EO

```python
import numpy as np

fair_method = "van"
lams = [0.0]
mode  = "eo"
name = "adult"
exps = [i for i in range(10)]
T = 0
for lam in lams:
    aps = []
    gap_logits = []
    gap_05s = []
    print("lam:", lam)
    for exp in exps:
        fname = f"results/Normal/{mode}/{fair_method}/Exp{exp}_{name}_{fair_method}_lam-{lam}_T-{T}.txt"
        with open(fname, "r") as f:
            ap = float(f.readline())
            gap_logit = float(f.readline())
            gap_05 = float(f.readline())
            aps.append(ap)
            gap_logits.append(gap_logit)
            gap_05s.append(gap_05)
    print(np.mean(aps))
#         print(np.mean(gap_logits))
    print(np.mean(gap_05s))
    
    
    
fair_method = "os"
lams = [0.0]
mode  = "eo"
name = "adult"
exps = [i for i in range(10)]
T = 0
for lam in lams:
    aps = []
    gap_logits = []
    gap_05s = []
    print("lam:", lam)
    for exp in exps:
        fname = f"results/Normal/{mode}/{fair_method}/Exp{exp}_{name}_{fair_method}_lam-{lam}_T-{T}.txt"
        with open(fname, "r") as f:
            ap = float(f.readline())
            gap_logit = float(f.readline())
            gap_05 = float(f.readline())
            aps.append(ap)
            gap_logits.append(gap_logit)
            gap_05s.append(gap_05)
    print(np.mean(aps))
#         print(np.mean(gap_logits))
    print(np.mean(gap_05s))
    
fair_method = "reweight"
lams = [0.0]
mode  = "eo"
name = "adult"
exps = [i for i in range(10)]
Ts = [0,1,2,10]
for T in Ts:
    for lam in lams:
        aps = []
        gap_logits = []
        gap_05s = []
        print("lam:", lam)
        for exp in exps:
            fname = f"results/Normal/{mode}/{fair_method}/Exp{exp}_{name}_{fair_method}_lam-{lam}_T-{T}.txt"
            with open(fname, "r") as f:
                ap = float(f.readline())
                gap_logit = float(f.readline())
                gap_05 = float(f.readline())
                aps.append(ap)
                gap_logits.append(gap_logit)
                gap_05s.append(gap_05)
        print(np.mean(aps))
    #         print(np.mean(gap_logits))
        print(np.mean(gap_05s))
```
